# Simple Animations in Jupyter Notebooks with `IPython.display`

This notebook illustrates steps you might take in figuring out how to do simple animations in jupyter notebooks using `IPython.display` functions.  The second to last code cell shows the best way to make an animation.  Run each code cell to see its effects.

```python
import numpy as np
import matplotlib.pyplot as plt
import IPython.display as ipd  # for display and clear_output
import time  # for sleep
```

```python
# linspace returns evenly space numbers, this is returning 100 numbers spaced evenly
# between -1 and 1
xs = np.linspace(-1, 1, 100)
# 10 nums spaced evenly between 0 and 10
for f in np.linspace(0, 10, 10):
    # plot sin of f*s numpy arrays
    plt.plot(xs, np.sin(f * xs))
```

```python
xs = np.linspace(-1, 1, 100)
for f in np.linspace(0, 10, 10):
    plt.figure()
    plt.plot(xs, np.sin(f * xs))
```

```python
xs = np.linspace(-1, 1, 100)
plt.figure()
for f in np.linspace(0, 10, 10):
    plt.plot(xs, np.sin(f * xs))
```

```python
xs = np.linspace(-1, 1, 100)
plt.figure()
for f in np.linspace(0, 10, 10):
    # clf() clears the plot !
    plt.clf()
    # only see the last one when this is run because that is the only one not cleared
    plt.plot(xs, np.sin(f * xs))
```

```python
xs = np.linspace(-1, 1, 100)
plt.figure()
for f in np.linspace(0, 10, 10):
    plt.clf()
    plt.plot(xs, np.sin(f * xs))
    # sleep in attempt to wait before next iteration is run and plot is cleared
    # but it happens so fast, jupyter actually doesn't display all of them
    time.sleep(0.2)  # 0.2 sec
```

```python
xs = np.linspace(-1, 1, 100)
fig = plt.figure()
# plots each one in its own graph
for f in np.linspace(0, 10, 10):
    plt.clf()
    plt.plot(xs, np.sin(f * xs))
    ipd.display(fig)
    time.sleep(0.2)  # 0.2 seconds
```

```python
xs = np.linspace(-1, 1, 100)
fig = plt.figure()
for f in np.linspace(0, 10, 10):
    plt.clf()
    plt.plot(xs, np.sin(f * xs))
    # clear_output allows for each plot to be plotted on the "same" graph
    # but the last one is displayed on a separate one
    ipd.clear_output(wait=True)
    ipd.display(fig)
    time.sleep(0.2)  # 0.2 seconds
```

```python
xs = np.linspace(-1, 1, 100)
fig = plt.figure()
for f in np.linspace(0, 10, 10):
    plt.clf()
    plt.plot(xs, np.sin(f * xs))
    ipd.clear_output(wait=True)
    ipd.display(fig)
    time.sleep(0.2)  # 0.2 seconds
ipd.clear_output(wait=True)
```

```python
xs = np.linspace(-1, 1, 100)
fig = plt.figure()
for f in np.linspace(0, 10, 10):
    # by not clearing the plot, we can see each function on the same graph incrementally
    #plt.clf()
    plt.plot(xs, np.sin(f * xs))
    ipd.clear_output(wait=True)
    ipd.display(fig)
    time.sleep(0.2)  # 0.2 seconds
ipd.clear_output(wait=True)
```

So, in general follow this template:

```python
...
fig = plt.figure()
for ... :
    plt.clf()
    # ... code to draw one frame
    ipd.clear_output(wait=True)
    ipd.display(fig)
    time.sleep(...)
ipd.clear_output(wait=True)
    
```
